```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial import ConvexHull
import folium
```

```python
from unicodedata import decimal


df = pd.read_csv("data/processed/kmeans.csv")

df = df.rename(columns = {'Unnamed: 0': 'id'})
```

```python
df[df.decimalLongitude < -100]
```

```python
K_clusters = range(1,10)

kmeans = [KMeans(n_clusters=i) for i in K_clusters]

Y_axis = df[['decimalLatitude']]

X_axis = df[['decimalLongitude']]

score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]

# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
from enum import unique


kmeans = KMeans(n_clusters = 3, init ='k-means++')
kmeans.fit(df[df.columns[1:3]]) # Compute k-means clustering.
df['cluster_label'] = kmeans.fit_predict(df[df.columns[1:3]])+1
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
labels = kmeans.predict(df[df.columns[1:3]]) # Labels of each point
df.head(10)
```

```python
df.plot.scatter(x = 'decimalLongitude', y = 'decimalLatitude', c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 1], centers[:, 0], c='black', s=200, alpha=0.5)
```

```python
import numpy as np
df['colors'] = np.where(
    df['cluster_label'] == 1, 'blue', np.where(
    df['cluster_label'] == 2 , 'red', 'green')) 
```

```python
m = folium.Map(zoom_start=4)
for index, row in df.iterrows():
    folium.CircleMarker([row['decimalLatitude'], row['decimalLongitude']],
                    radius=3, color=row['colors']).add_to(m)


m
```
